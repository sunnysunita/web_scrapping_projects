```
import requests
```

```
import json
```

```
base_url = "https://jobs.github.com/positions.json"
```

```
response = requests.get(base_url, params={"description":"java script","page":2})
print(response.status_code)
print(response.ok)
```

```
resp_data = response.json()
print(json.dumps(resp_data, indent = 4))
print(type(resp_data))
```

```
print(len(resp_data))
```

```
import pandas as pd
```

```
job = pd.DataFrame(resp_data)
```

```
page = 5
job_list = []
for i in range(1,page+1):
  resp_data = requests.get(base_url, params={"description":"java script","page":i}).json()
  if len(resp_data) == 0:
    break
  else:
    job_list.append(resp_data)

print(len(job_list))
```
