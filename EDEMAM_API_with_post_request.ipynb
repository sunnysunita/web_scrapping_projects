```
import requests
import json
import pandas as pd
```

```
app_id = "0a5f77b8"
app_key = "05f33c28a81b4e1347a5082a77fb8b78"
```

```
base_url = "https://api.edamam.com/api/nutrition-details"
```

```
authenticate_url = base_url + "?app_id=" + app_id + "&app_key=" + app_key
```

```
header = {
    "Content-Type": "application/json"
}
```

```
recipe = {
    "title":"Cappauccino",
    "ingr":["18g ground espresso","150ml milk"]
}
```

```
response = requests.post(authenticate_url, headers = header, json = recipe)
```

```
resp_data = response.json()
```

```
print(type(resp_data))
```
