```
import requests
```

```
from bs4 import BeautifulSoup
```

```
base_url = "https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/2/"
```

```
response = requests.get(base_url)
response.status_code
```

```
soup = BeautifulSoup(response.content, "lxml")
file = open("rotten.html","wb")
file.write(soup.prettify("utf-8"))
soup
```

```
outer_block = soup.find_all("div", attrs = {"class":"col-sm-18 col-full-xs countdown-item-content"})
outer_block
```

```
list_movies = []
for e in outer_block:
  movie_dic = {}
  temp = e.find("h2")
  movie_dic["Movie_Name"] = temp.find("a").string
  movie_dic["Year_Release"] = temp.find("span", attrs={"class":"subtle start-year"}).string
  movie_dic["Rating"] = temp.find("span", attrs={"class":"tMeterScore"}).string
  movie_dic["Critics_Consensus"] = e.find("div", attrs = {"class":"info critics-consensus"}).text[19:]
  movie_dic["Synopsis"] = e.find("div", attrs = {"class":"info synopsis"}).text[9:]
  movie_dic["Starring"] = e.find("div", attrs = {"class":"info cast"}).text.strip()[9:]
  movie_dic["Directed_By"] = e.find("div", attrs = {"class":"info director"}).text.strip()[12:]
  list_movies.append(movie_dic)
```
